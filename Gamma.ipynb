```python
import sys
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import numpy as np
import cv2
```

```python
#Load image
img = cv2.imread("im6.png", cv2.IMREAD_GRAYSCALE)
fig1=plt.figure(figsize=(15,8))
plt.imshow(img,cmap="gray",vmin=0,vmax=255)
plt.axis("off")
#TODO extraer los canales de color
print(img.shape)
```

```python
gamma=1

img=img.astype(float)
inmax=np.amax(img)

c=255/(np.log(1+inmax))
Iout = img/inmax
Iout = Iout**gamma
Iout = Iout*255

Iout = Iout.astype(np.uint8)    #Conversion de ndarray a int

fig3 = plt.figure(figsize=(15,8))
a = fig3.add_subplot(1,2,1)
a.axis('off')
plt.imshow(Iout,cmap="gray",vmin=0,vmax=255)
a.set_title("Contraste gamma")

#Imagen original
img=img.astype(np.uint8)
a = fig3.add_subplot(1,2,2)
a.axis('off')
plt.imshow(img,cmap="gray",vmin=0,vmax=255)
a.set_title("Original")
```
