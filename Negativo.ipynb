```python
import sys
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import numpy as np
```

```python
img = plt.imread("im4.png")
fig1=plt.figure(figsize=(15,8))
plt.imshow(img,cmap="gray",vmin=0,vmax=255)
plt.axis("off")
```

```python
inmax=np.amax(img)
inmin=np.amin(img)

hOut = 255
m=255/(inmax-inmin)
b=-m*inmin
Iout=img*m+b
Iout=Iout.astype(float)


#Original
fig1=plt.figure(figsize=(15,8))
c = fig1.add_subplot(1,3,1)
plt.imshow(img,cmap="gray",vmin=0,vmax=255)
plt.axis("off")
c.set_title("Original")

print(type(Iout))
Iout=Iout.astype(np.uint8)  #Conversion de ndarray a int
a = fig1.add_subplot(1,3,2)
plt.imshow(Iout,cmap='gray',vmin=0,vmax=255)
plt.axis('off')
a.set_title("Impro")

Iout = 255-Iout
Iout=Iout.astype(np.uint8)  #Conversion de ndarray a int
b = fig1.add_subplot(1,3,3)
plt.imshow(Iout,cmap='gray',vmin=0,vmax=255)
plt.axis('off')
b.set_title("Negativo")
```

```python
#Load image
img = plt.imread("im9.jpg")
fig1=plt.figure(figsize=(15,8))
plt.imshow(img,cmap="gray",vmin=0,vmax=255)
plt.axis("off")
```

```python
img=img.astype(float)
inmax=np.amax(img)

c=255/(np.log(1+inmax))

Iout = np.log(img+1)
Iout = c*Iout
Iout = Iout.astype(np.uint8)

fig3 = plt.figure(figsize=(15,8))
a = fig3.add_subplot(1,2,1)
a.axis('off')
plt.imshow(Iout,cmap="gray",vmin=0,vmax=255)
a.set_title("Correcccion logarítmica")

#Imagen original
img=img.astype(np.uint8)
a = fig3.add_subplot(1,2,2)
a.axis('off')
plt.imshow(img,cmap="gray",vmin=0,vmax=255)
a.set_title("Original")
```

```python
#Histograma de la imagen original contra la corrección logarítmica
fig2 = plt.figure(figsize = (15,8))
a = fig2.add_subplot(1, 2, 1)
arr = np.array(img)
arr = arr.ravel()
plt.hist(arr, 255)
plt.xlim([0,255])


b = fig2.add_subplot(1, 2, 2)
arr = np.array(Iout)
arr = arr.ravel()
plt.hist(arr, 255)
plt.xlim([0,255])
```
