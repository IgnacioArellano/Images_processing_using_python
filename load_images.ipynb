```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r'C:\Users\coyul\Documents\GitHub\PDI\im4.png', cv2.IMREAD_GRAYSCALE)
```

```python
roi=cv2.selectROI('select',img)
cv2.destroyWindow('select')
```

```python
ROI_image=img[roi[1]:roi[1]+roi[3], roi[0]:roi[0]+roi[2]]
```

```python
fig1 = plt.figure(figsize = (6,6))
a = fig1.add_subplot(1, 1, 1)
plt.imshow(ROI_image, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
a.set_title('Image');
```
