```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread(r"im1.jpg",cv2.IMREAD_GRAYSCALE)
equ = cv2.equalizeHist(img)     #equqlizacin de histograma
img = cv2.hconcat([img,equ])
plt.imshow(img)
plt.axis("off")
```

```python
I = cv2.imread('im11.jpg',cv2.IMREAD_GRAYSCALE)
# Open CV 
If = cv2.blur(I,(5, 5))     #desenfoque (promedio)
imgM = cv2.medianBlur(I,7)
imgG = cv2.GaussianBlur(I,(9,9),cv2.BORDER_DEFAULT)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
ax[0, 0].imshow(I, cmap='gray')
ax[0, 0].set_title('Original image')

ax[0, 1].imshow(If, cmap='gray')
ax[0, 1].set_title('Blur')

ax[1, 0].imshow(imgM, cmap='gray')
ax[1, 0].set_title('Median Blur')

ax[1, 1].imshow(imgG, cmap='gray')
ax[1, 1].set_title('Gausian blur')

for a in ax.ravel():
    a.axis('off')

plt.tight_layout()

plt.show()
```

```python
I = cv2.imread('im10.jpg',cv2.IMREAD_GRAYSCALE)
# Open CV 
If = cv2.blur(I,(5, 5))     #desenfoque (promedio)

imgDenoised = cv2.fastNlMeansDenoisingMulti(I, 2, 5, None, 4, 7, 35)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
ax[0, 0].imshow(I, cmap='gray')
ax[0, 0].set_title('Original image')

ax[0, 1].imshow(If, cmap='gray')
ax[0, 1].set_title('Fast Denoising')
```

```python
from skimage.filters import rank,gaussian
from skimage.morphology import disk,rectangle,square
from skimage.restoration import denoise_nl_means,estimate_sigma
import scipy.ndimage as ndi
```

```python
I = cv2.imread('imgRuidoGausiano.jfif',cv2.IMREAD_GRAYSCALE)
I=I.astype(float)
sigma_est=np.mean(estimate_sigma(I))

In=denoise_nl_means(I,h=1.15*sigma_est,fast_mode=False,patch_size=5,patch_distance=8)
#dst = cv2.fastNlMeansDenoisingColored(I,None,10,10,7,21)
I=I.astype(np.uint8)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))
ax[0].imshow(I, cmap='gray')
ax[0].set_title('Original image')

ax[1].imshow(In, cmap='gray')
ax[1].set_title('no local means')

```
