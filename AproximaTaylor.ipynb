```python
import numpy as np
import matplotlib.pyplot as plt
# if using a jupyter notebook
# %matplotlib inline 
```

```python
#Define sin x
x = np.arange(-6*np.pi,6*np.pi,0.1)   # start,stop,step
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
#Graficar serie
N=15 #Máximo N
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y=x*0

for n in range(0,N):
    y = np.add(y,((-1)**n)*(x**(2*n+1))/(np.math.factorial(2*n+1)))

#n=0
#y = ((-1)**n)*(x**(2*n+1))/(np.math.factorial(n))

plt.plot(x,y)
plt.show()
```
