### Load and show images using matplotlib

```python
import sys 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = plt.imread('im1.jpg')
Size=img1.shape
fig1=plt.figure(figsize = (4,4))
plt.imshow(img1, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
plt.title(['Color image of 8bits and size = '+str(Size)]);
```

```python
img2 = plt.imread('/home/pds/Documents/GitHub/Images_processing_using_python/im2.jpeg')
Size=img2.shape
fig2=plt.figure(figsize = (4,4))
plt.imshow(img2, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
plt.title(['Gray image of 8bits and size = '+str(Size)]);
```

```python
img3 = plt.imread('/home/pds/Documents/GitHub/Images_processing_using_python/im3.tif')
Size=img3.shape
fig3=plt.figure(figsize = (4,4))
plt.imshow(img3, cmap='gray',vmin=0,vmax=2**16)
plt.axis('off');
plt.title(['Gray image of 16bits and size = '+str(Size)]);
```

### Show image as surface

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib qt
```

```python
xx, yy = np.mgrid[0:img3.shape[0], 0:img3.shape[1]]
fig4 = plt.figure()
ax = fig4.gca(projection='3d')
ax.plot_surface(xx, yy, img3, cmap=plt.cm.hot)
plt.show()
```

### Show images at subplot

```python
fig5 = plt.figure(figsize = (15,8))

a = fig5.add_subplot(1, 3, 1)
plt.imshow(img1, cmap='gray',vmin=0,vmax=2**8)
plt.axis('off');
a.set_title('Image 1')

b = fig5.add_subplot(1, 3, 2)
b.set_title('Image 2')
plt.imshow(img2, cmap='gray',vmin=0,vmax=2**8)
plt.axis('off')

c = fig5.add_subplot(1, 3, 3)
c.set_title('Image 3')
plt.imshow(img3, cmap='gray',vmin=0,vmax=2**16)
plt.axis('off');
```
